```python
import hashlib
from pathlib import Path
from markdown_it import MarkdownIt
from markdown_it.token import Token
from PIL import Image

# ---------- Stub caption/OCR -----------
def generate_caption(image_path: Path) -> str:
    # TODO: replace with BLIP-2 model call
    return f"[CAPTION] Placeholder caption for {image_path.name}"

def run_ocr(image_path: Path) -> str:
    # TODO: replace with docTR or Tesseract OCR
    return f"[OCR] Extracted text from {image_path.name}"

# ---------- Data containers ------------
# 

class Chunk:
    def __init__(self, chunk_id, doc_id, node_type, raw, plain, section_path, offsets):
        self.chunk_id = chunk_id        # Unique ID for this chunk (local to your pipeline)
        self.doc_id = doc_id            # ID of the whole document/report this chunk came from
        self.node_type = node_type      # Type of markdown block (heading, paragraph, table, list, etc.)
        self.raw_markdown = raw         # Original markdown text for this chunk (with formatting)
        self.plain_text = plain         # Cleaned plain-text version (no markdown syntax)
        self.section_path = section_path# Hierarchical path of headings leading to this chunk (e.g. "Financials/NAV")
        self.offsets = offsets          # (start_line, end_line) positions in the original file for provenance
        self.image_ids = []             # List of image IDs linked to this chunk (if any images appear inside it)

    def build_index_text(self, image_refs):
        parts = [self.plain_text]
        for img in image_refs:
            if img.alt_text: parts.append(f"[IMG ALT] {img.alt_text}")
            if img.caption_text: parts.append(f"[IMG CAPTION] {img.caption_text}")
            if img.ocr_text: parts.append(f"[IMG OCR] {img.ocr_text}")
        return "\n".join(parts)

class ImageRef:
    def __init__(self, image_id, doc_id, chunk_id, section_path, path, alt_text):
        self.image_id = image_id
        self.doc_id = doc_id
        self.chunk_id = chunk_id
        self.section_path = section_path
        self.path = path
        self.alt_text = alt_text
        self.caption_text = None
        self.ocr_text = None
        self.hash_md5 = None
        self.width = None
        self.height = None

# ---------- Utilities ------------------
def hash_file(path: Path) -> str:
    md5 = hashlib.md5()
    with open(path, "rb") as f:
        for chunk in iter(lambda: f.read(8192), b""):
            md5.update(chunk)
    return md5.hexdigest()

# ---------- Main ingest ----------------
def parse_markdown(md_path: Path, doc_id: str):
    md = MarkdownIt("commonmark").enable("table")  # enable GFM table ext if needed
    text = md_path.read_text(encoding="utf-8")

    tokens = md.parse(text)
    section_stack = []
    chunks, images = [], []
    chunk_id_counter, img_id_counter = 1, 1

    for tok in tokens:
        # Headings update section_path
        if tok.type == "heading_open":
            level = int(tok.tag[1])  # e.g. "h2" -> 2
            # Pop deeper/equal levels
            section_stack = section_stack[: level - 1]
        elif tok.type == "inline" and tokens[tokens.index(tok)-1].type.startswith("heading_open"):
            section_stack.append(tok.content)

        # Paragraphs / tables / etc.
        if tok.type in ["paragraph_open", "table_open", "bullet_list_open", "ordered_list_open"]:
            node_type = tok.type.split("_")[0]
            
            
            # Collect its inline text
            inline_text = ""
            raw = ""
            # look ahead
            for t in tokens[tokens.index(tok)+1:]:
                if t.type.endswith("_close") and t.level == tok.level:
                    break
                raw += t.content
                if t.type == "inline":
                    inline_text += t.content
            chunk = Chunk(
                chunk_id=f"chunk_{chunk_id_counter}",
                doc_id=doc_id,
                node_type=node_type,
                raw=raw,
                plain=inline_text,
                section_path="/".join(section_stack),
                offsets=(tok.map[0], tok.map[1]) if tok.map else None,
            )
            chunks.append(chunk)
            chunk_id_counter += 1

        # Images
        if tok.type == "inline" and tok.children:
            for child in tok.children:
                if child.type == "image":
                    img = ImageRef(
                        image_id=f"img_{img_id_counter}",
                        doc_id=doc_id,
                        chunk_id=f"chunk_{chunk_id_counter-1}" if chunks else None,
                        section_path="/".join(section_stack),
                        path=str((md_path.parent / child.attrs["src"]).resolve()),
                        alt_text=child.attrs.get("alt", "")
                    )
                    p = Path(img.path)
                    if p.exists():
                        img.hash_md5 = hash_file(p)
                        try:
                            im = Image.open(p)
                            img.width, img.height = im.size
                        except Exception:
                            pass
                        # Add caption/OCR
                        img.caption_text = generate_caption(p)
                        img.ocr_text = run_ocr(p)
                    images.append(img)
                    img_id_counter += 1
                    if chunks: chunks[-1].image_ids.append(img.image_id)

    return chunks, images
```

```python
md_file = Path("out/HGGC_Fund_IV_-_Q1_2023_Report-with-image-refs.md")
chunks, images = parse_markdown(md_file, doc_id="HGGC_Fund_IV_-_Q1_2023_Report")

for c in chunks:
    imgs = [i for i in images if i.chunk_id == c.chunk_id]
    print("Chunk:", c.section_path, c.build_index_text(imgs))
for img in images:
    print("Image:", img.path, img.caption_text, img.ocr_text)
```
